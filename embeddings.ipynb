```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="this is something on OPENAI embedding"
query_result=embeddings.embed_query(text)
query_result
```

```python
from langchain_community.document_loaders import TextLoader
text=TextLoader('speech.txt').load()
from langchain_text_splitters import RecursiveCharacterTextSplitter
Splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_doc=Splitter.split_documents(text)
final_doc
```

```python
import chromadb

from langchain_community.vectorstores import Chroma
```

```python
db=Chroma.from_documents(final_doc,embeddings)
db
```

```python
#retrieving the results from query vectorstore db
query="It will be all the easier"
retrieved_results=db.similarity_search(query)
```
