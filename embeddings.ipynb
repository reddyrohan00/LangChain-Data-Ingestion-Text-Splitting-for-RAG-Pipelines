```python
from langchain_community.document_loaders import TextLoader
text=TextLoader('speech.txt').load()
from langchain_text_splitters import RecursiveCharacterTextSplitter
Splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_doc=Splitter.split_documents(text)
final_doc
```

```python
import chromadb

from langchain_community.vectorstores import Chroma
```

```python
db=Chroma.from_documents(final_doc,embeddings)
db
```

```python
#retrieving the results from query vectorstore db
query="It will be all the easier"
retrieved_results=db.similarity_search(query)
```

```python
print(retrieved_results)
```

```python
from langchain.embeddings import OllamaEmbeddings
ollamaembeddings=OllamaEmbeddings(model="nomic-embed-text")
Text="Initial steps of langchain"
query_result=ollamaembeddings.embed_query(text=Text)
```

```python
print(query_result)
```

```python
Text2="first steps of learning langchain"
query_result2=ollamaembeddings.embed_query(text=Text2)
print(query_result2)
```

```python
# Convert to numpy arrays and reshape (required for sklearn)
import sklearn,numpy as np
from sklearn.metrics.pairwise import cosine_similarity
vec1 = np.array(query_result).reshape(1, -1)
vec2 = np.array(query_result2).reshape(1, -1)

similarity = cosine_similarity(vec1, vec2)[0][0]

print(f"Cosine similarity between text1 and text2: {similarity:.4f}")
```
